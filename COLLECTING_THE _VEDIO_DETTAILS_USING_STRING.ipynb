```python
import pandas as pd
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError

# Your API key
api_key = 'AIzaSyB-wavsGsS_1Nekw2zux_2Gmlugop6Uous'

# Create a YouTube service object
youtube = build('youtube', 'v3', developerKey=api_key)

# Define the search query
search_query = '(Electronic nicotine delivery systems or ENDS)'

video_data = []

try:
    # Call the search.list method to retrieve video details
    search_response = youtube.search().list(
        q=search_query,
        part='snippet',
        maxResults=200  # Change this number as per your requirement
    ).execute()

    # Extract video details
    for search_result in search_response.get('items', []):
        if search_result['id']['kind'] == 'youtube#video':
            video_id = search_result['id']['videoId']
            video_response = youtube.videos().list(
                part='snippet, statistics, contentDetails',
                id=video_id
            ).execute()

            video_stats = video_response['items'][0]['statistics']
            video_snippet = video_response['items'][0]['snippet']
            video_duration = video_response['items'][0]['contentDetails']['duration']
            video_link = f"https://www.youtube.com/watch?v={video_id}"
            comment_count = video_stats.get('commentCount', 0)

            video_data.append({
                'Title': video_snippet['title'],
                'Views': video_stats.get('viewCount', 0),
                'Likes': video_stats.get('likeCount', 0),
                'Dislikes': video_stats.get('dislikeCount', 0),
                'Uploaded Date': video_snippet['publishedAt'],
                'Duration': video_duration,
                'Video Link': video_link,
                'Comment Count': comment_count
            })

except HttpError as e:
    print("An HTTP error occurred:", e.content)
except Exception as ex:
    print("An error occurred:", str(ex))

# Create a DataFrame from the video data
df = pd.DataFrame(video_data)

df
```
